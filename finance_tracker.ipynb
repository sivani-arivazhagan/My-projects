```python
# defining my spending categories
categories = ("food","travel","shopping")
print("Categories:", categories)
```

```python
# caclulating pay/income

# defining my hourly rate
hourly_rate = 15

# calculate hours for 2 different shifts
shift_1 = 4*4 # 4 hours each for 4 days
shift_2 = 2*7 # 7 hours each for 2 days

# calculate total working hours
total_working_hours = shift_1 + shift_2
print("Total working hours:", total_working_hours)
total_income = hourly_rate*total_working_hours
print("total income:", total_income)
```

```python
total_income = 450.00
# dictionary
spending = {
    "food": 110,
    "travel": 30,
    "shopping": 90
}
print("Cateogory breakdown")
for category, amount in spending.items():
    print(f"{category}: ${amount}")

# calculate total spending
total_spent = sum(spending.values())

# calculate savings
remaining_balance = total_income - total_spent

print(f"TOTAL SPENT:      ${total_spent:.2f}")
print(f"REMAINING BUDGET: ${remaining_balance:.2f}")
```

```python
# data visualisation
import matplotlib.pyplot as plt

# defining data
labels = ["Food", "Travel", "Shopping", "Savings"]
amounts = (110, 30, 90, 220)

# creating a pie chart
plt.pie(amounts, labels=labels, autopct="%1.1f%%")
plt.title('My monthly spending breakdown')
plt.axis('equal') # this makes it a circle

```
